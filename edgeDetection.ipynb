```python
import cv2 as cv
import matplotlib.pyplot as plt
imgPath = r"C:\Users\mynam\Documents\opencvPython\360_F_268531917_ZffyVqTmfTE7gCYz7qOulpRSxVmZfwa5.jpg"
img = cv.imread(imgPath) 
grayImage = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(grayImage)

blurredImage = cv.GaussianBlur(grayImage, (3,3), 0) 
plt.imshow(blurredImage)
```

```python
sobelx = cv.Sobel(src=blurredImage, ddepth=cv.CV_64F, dx=1, dy=0, ksize=5)
sobely = cv.Sobel(src=blurredImage, ddepth=cv.CV_64F, dx=0, dy=1, ksize=5)
sobelxy = cv.Sobel(src=blurredImage, ddepth=cv.CV_64F, dx=1, dy=1, ksize=5) 
```
